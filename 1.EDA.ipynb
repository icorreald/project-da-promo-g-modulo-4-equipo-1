```python
df2 = pd.read_csv('recursos-humanos.csv', index_col=0)
```

```python
def exploracion_basica(lista_df):
    for df in lista_df:
       display(df.head())
       print(df.shape)
       print('')
       print(df.isnull().sum()/df.shape[0]*100)
       print('')
       for columna in df.columns:
        print(df[columna].duplicated().sum()/df.shape[0]*100)
       print('')
       print(df.dtypes)
       for columna in df.select_dtypes(include='O'):
        print(df[columna].value_counts())
        print('')
```

```python
exploracion_basica([df2])
```
